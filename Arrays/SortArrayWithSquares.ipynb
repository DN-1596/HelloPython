```python
# Problem Description

# Given a sorted array A containing N integers both positive and negative.

# You need to create another array containing the squares of all the elements in A and return it in non-decreasing order.

# Try to this in O(N) time.


# Problem Constraints
# 1 <= N <= 10^5.

# -10^3 <= A[i] <= 10^3



# Input Format
# First and only argument is an integer array A.



# Output Format
# Return a integer array as described in the problem above.



# Example Input
# Input 1:

#  A = [-6, -3, -1, 2, 4, 5]
# Input 2:

#  A = [-5, -4, -2, 0, 1]


# Example Output
# Output 1:

#  [1, 4, 9, 16, 25, 36]
# Output 2:

#  [0, 1, 4, 16, 25]
```

```python
class Solution:
    def solve(self,A):
        
        A1 = []
        A2 = []
        
        Res = []
        
        for i in A:
            
            if i < 0 : 
                A1.append(i**2)
            else :
                A2.append(i**2)
        
        A1.reverse()
        
        i = 0 
        j = 0
        
        while i < len(A1) or j < len(A2) : 
            
            
            if i >= len(A1): 
                Res.append(A2[j])
                j+=1
            elif j >= len(A2): 
                Res.append(A1[i])
                i+=1
            elif A1[i] < A2[j] : 
                Res.append(A1[i])
                i+=1
            else :
                Res.append(A2[j])
                j+=1
        
        return Res

print(Solution().solve([-6, -3, -1, 2, 4, 5]))
            
            
            
        
```
