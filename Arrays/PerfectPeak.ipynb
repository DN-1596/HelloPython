```python
class Solution:
    
    def perfectPeak(self,A):
        
        N = len(A)
        
        maxArray = [0]*N
        maxEle = -1
        minArray = [0]*N 
        minEle = 10**9
        
        for i in range(0,N):
            
            maxArray[i] = maxEle 
            maxEle = max(maxEle,A[i])
            
            i = N-1-i 
            
            minArray[i] = minEle
            minEle = min(minEle,A[i]) 
        
        for i in range(1,N-1):
            
            if maxArray[i] < A[i] and A[i] < minArray[i]:
                return 1
        
        return 0
            

print(Solution().perfectPeak(([9895, 30334, 17674, 23812, 20038, 25668, 6869, 1870, 4665, 27645, 7712, 17036, 31323, 8724, 28254, 28704, 26300, 25548, 15142, 12860, 19913, 32663, 32758 ])))
            
            
            
            
        
```
