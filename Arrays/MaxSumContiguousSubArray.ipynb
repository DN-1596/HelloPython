```python
class Solution:
    
    def maxSubArray(self,A):
    
        maxTotal = -1001
        total = 0;
        
        for i in range(len(A)) :
            total += A[i] 
            maxTotal = max(maxTotal,total)
            
            if total < 0 : 
                total = 0
        return maxTotal
    
    

print(Solution().maxSubArray([-2, 1, -3, 4, -1, 2, 1, -5, 4]))
    
            
                
        
        
        
        
    
```

```python
#BRUTE FORCE
class Solution:
    def maxSubArray(self,A):
        
        maxTot = -1001
        
        for i in range(len(A)):
            for j in range(i,len(A)):
                maxTot = max(maxTot,sum(A[i:j+1]))
            
        return maxTot

print(Solution().maxSubArray([2,-1,-2,4,3]))
```
