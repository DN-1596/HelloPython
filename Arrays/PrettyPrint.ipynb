```python
for i in range(4,0,-1):
    print(i)
```

```python
class Solution:
    def prettyPrint(self,n):
        resLength = 2*n-1
        res = [[0 for i in range(resLength)] for i in range(resLength)]
        start = 0
        end = resLength-1
        
        for val in range(n,0,-1):
            
            for j in range(start,end+1):

                res[start][j] = val
                res[end][j]   = val
                res[j][start] = val
                res[j][end]   = val
 
            
            start+=1
            end-=1
                   
        
        return res

res = Solution().prettyPrint(4)

for row in res:
    print(row)
```
