```python
# Implement the next permutation, which rearranges numbers into the numerically next 
# greater permutation of numbers for a given array A of size N.

# If such an arrangement is not possible, it must be rearranged in the lowest possible order 
# i.e.,
# sorted in ascending order.

# Note:
# The replacement must be in-place, do not allocate extra memory.
# DO NOT USE LIBRARY FUNCTION FOR NEXT PERMUTATION. 
# Use of Library functions will disqualify your submission retroactively 
# and will give you penalty points.


# Problem Constraints
# 1 <= N <= 5e5
# 1 <= A[i] <= 1e9


# Input Format
# The first and the only argument of input has an array of integers, A.


# Output Format
# Return an array of integers, representing the next permutation of the given array.
```

```python
class Solution:
    def nextPermutation(self,A): 
        
        n = len(A)
        orgArray = self.getNumString(A)
        minPermut = [0]*n
        
        for i in range(n):
            B = [0]*n
            
            for i in range(n):
                B[i] = A[i]    
            
            for j in range(n): 
                
                if B[j] > B[i]:
                    
                    B[j] , B[i] = B[i] , B[j]
                    newA = self.getNumString(B)
                    
                    if orgArray < newA and self.getNumString(minPermut) > newA:
                        
                        for i in range(n):
                            minPermut[i] = B[i]    
                            
            
        return minPermut if self.getNumString(minPermut) > orgArray else sorted(A)
    
    def getNumString(self,A) :
        string = ""
        for i in range(len(A)): 
            string+=str(A[i])
        return string

s = Solution()

print(s.nextPermutation([20, 113, 50]))
        
        
```
